## 3. Python-Liste versus NumPy-Array
### a) Große Python-Liste erstellen


```python
size = 10_000_0000  # Anzahl der Elemente (kann verändert werden)
python_list = list(range(size))
```

### b) NumPy-Array erstellen

```python
numpy_array = np.arange(size)
```

### c) Quadrieren + Zeitmessung

```python
# Python-Liste (mit Schleife)
start = time.time()

python_result = []
for x in python_list:
    python_result.append(x ** 2)

python_time = time.time() - start


# NumPy-Array (vektorisiert)
start = time.time()

numpy_result = numpy_array ** 2

numpy_time = time.time() - start
```

### d) Ergebnisse Anzeigen

```python
print("Zeit Python-Liste:", python_time, "Sekunden")
print("Zeit NumPy-Array:", numpy_time, "Sekunden")
```

## 4. NumPy: Auf Array-Elemente zugreifen (Indexing)

### a) Zugriff auf ein Element an Stelle X

Der Index beginnt bei Python immer bei 0.

```python
arr = np.array([10, 20, 30, 40, 50])

print(arr[2])  # Element an Position 2
```

### b) Intervalle auswählen (Slicing)

```python
print(arr[1:4])
```

### c) Jeden 2. Wert auf Null setzen

::2 bedeutet:  vom Anfang bis Ende in 2er-Schritten

Dadurch wird jedes zweite Element ausgewählt.`

```python
arr[::2] = 0
print(arr)
```

## 5. Pandas



### a) Daten einlesen

```python
data = {
    "Lernstunden_pro_Woche": [5, 10, 2, 15, 8, 12, 4, 9, 11, 3],
    "Anwesenheit_in_Prozent": [60, 95, 50, 100, 80, 90, 55, 85, 92, 40],
    "Pruefungsdauer_in_Minuten": [30, 45, 25, 50, 35, 40, 28, 38, 42, 20],
    "Bestanden": [0, 1, 0, 1, 1, 1, 0, 1, 1, 0]  # 1 = bestanden, 0 = durchgefallen
}

df = pd.DataFrame(data)
df.head()
```

### b) Daten analysieren


```python
# Anzeige der Spaltenbezeichnungen
print('Spalten: ', df.columns)

# Index - Zeilen haben typischerweise einen numerischen Index, können aber z.B. auch vom Typ String sein
print(df.index)
```

```python
# Übersicht und statistische Werte
df.describe()
```

```python
# Überblick über Datentypen und fehlende Werte
df.info()
```

### c) Filtern und Manipulieren

```python
# Filter: Nur bestandene Studenten
df[df["Bestanden"] == 1]

# Filter: Studenten mit mehr als 10 Lernstunden
print ([df["Lernstunden_pro_Woche"] > 10])

# Hypothese prüfen: Mehr Lernen → besseres Bestehen?
df.groupby("Bestanden")["Lernstunden_pro_Woche"].mean()

# Durchschnittliche Anwesenheit je Bestehensstatus
df.groupby("Bestanden")["Anwesenheit_in_Prozent"].mean()

```
